```python
import scanpy as sc
import pertpy as pt
import random
```

```python
adata = sc.read_h5ad("/Users/mamba/Downloads/compositional.h5ad")
```

```python
sccoda_model = pt.tl.Sccoda()
sccoda_data = sccoda_model.load(
    adata,
    type="sample_level",
)
```

```python
sccoda_data = sccoda_model.prepare(
    sccoda_data,
    formula="AGE",
    reference_cell_type="automatic",
)
sccoda_model.run_nuts(sccoda_data)
sccoda_model.summary(sccoda_data)
```

```python
sccoda_data = sccoda_model.prepare(
    sccoda_data,
    formula="age_group",
    reference_cell_type="automatic",
)
sccoda_model.run_nuts(sccoda_data)
sccoda_model.summary(sccoda_data)
```

```python
# This is the crucial part: we need to normalize the age covariate for the model to work
age = sccoda_data["coda"].obs["AGE"]
age_normalized = (age - age.min()) / (age.max() - age.min())
sccoda_data["coda"].obs["age_normalized"] = age_normalized

sccoda_data = sccoda_model.prepare(
    sccoda_data,
    formula="age_normalized",
    reference_cell_type="automatic",
)
sccoda_model.run_nuts(sccoda_data)
sccoda_model.summary(sccoda_data)
```
